```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/ksrivastava/Downloads/HistoricalData_1706977444449.csv")
```

```python
# convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# filter the dataframe to include 2017-12-07 to 2018-06-01
df = df[(df['Date'] >= '2017-12-07') & (df['Date'] <= '2018-06-01')]
```

```python
df
```

```python
newsdf = pd.read_csv("/Users/ksrivastava/Desktop/NLP/Datasets/news_data_labelled.csv")
newsdf['Date'] = pd.to_datetime(newsdf['Date'])
```

```python
# Step 2: Aggregate the data
articles_per_day = newsdf.groupby('Date')['Text'].count()

# Create a full date range
date_range = pd.date_range(start='2017-12-07', end='2018-06-01')

# Convert the date_range to a DataFrame
date_df = pd.DataFrame(date_range, columns=['Date'])

# Set the Date as the index for easy merging
date_df.set_index('Date', inplace=True)
articles_per_day = articles_per_day.reindex(date_df.index, fill_value=0)

# Step 3: Plot the data
articles_per_day.plot(kind='bar', figsize=(30, 20))

plt.title('Number of Articles per Day')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.xticks(rotation=90)  # Rotate date labels to avoid overlap
plt.tight_layout()  # Adjust layout to make room for the rotated date labels
plt.show()
```

```python
# Graph number of articles against the stock price
# Plot the stock price data
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Close/Last', color=color)
ax1.plot(df['Date'], df['Close/Last'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for the number of articles
ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('Number of Articles', color=color)
ax2.plot(articles_per_day.index, articles_per_day, color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Title and layout
plt.title('S&P 500 Close Price and Number of Articles per Day')
fig.tight_layout()

plt.show()
```
